```python
import numpy as np
import pandas as pd
import site
import sys
site.addsitedir(r'C:\Users\pcallahan\Desktop\school\580\gun-violence-data-master\p_callahan')
import load


law_tbl = pd.read_csv('law_tbl.csv')
print("imports completed \n")
df = pd.read_csv(r'C:\Users\pcallahan\Desktop\school\580\gun-violence-data-master\p_callahan\stage3.csv')
state_abrs = pd.read_csv(r'C:\Users\pcallahan\Desktop\school\580\gun-violence-data-master\p_callahan\state_abrs.csv')
df = pd.merge(df,state_abrs, on='state')
cico_dat = pd.read_csv(r'C:\Users\pcallahan\Desktop\school\580\gun-violence-data-master\p_callahan\cico_dat.csv')
print("city-county data loaded \n")
cico_dat['inj_qnt'] = cico_dat['n_injured']/cico_dat['population']
cico_dat['kill_qnt'] = cico_dat['n_killed']/cico_dat['population']
cico_dat['rpt_qnt'] = cico_dat['n_rpts']/cico_dat['population']
```

```python
state_dat_aves = cico_dat.groupby(['state_abr'])[['inj_qnt','kill_qnt','rpt_qnt']].mean()
state_dat_aves.reset_index(inplace=True)
state_dat_sums = cico_dat.groupby(['state_abr'])[['n_injured','n_killed','n_rpts']].sum()
state_dat_sums.reset_index(inplace=True)

state_dat = pd.merge(state_dat_aves, state_dat_sums, on=['state_abr'])

abrs = pd.read_csv(r'C:\Users\pcallahan\Desktop\school\580\gun-violence-data-master\p_callahan\state_abrs.csv')
state_dat = pd.merge(state_dat, abrs, on='state_abr')
vsis = pd.read_csv(r'C:\Users\pcallahan\Desktop\school\580\gun-violence-data-master\p_callahan\vega_state_ids.csv')
state_dat = pd.merge(state_dat, vsis, on='state')
state_dat
```

```python
# Get US State background from JSON data provided by Altair/Vega Datasets
import altair as alt
from vega_datasets import data
states = alt.topo_feature(data.us_10m.url, feature='states')
alt.data_transformers.enable('json') 

rq_map = alt.Chart(states).mark_geoshape().encode(
    color='rpt_qnt:Q',
    tooltip = ['state:N','rpt_qnt:Q']
).transform_lookup(
    lookup='id',
    from_=alt.LookupData(state_dat, 'vega_state_id', ['rpt_qnt','state'])
).project(
    type='albersUsa'
).properties(
    title = 'Ave Gun Crime Report Quotient by State (All)',
    width=800,
    height=450
)
rq_map
```

```python
import altair as alt
from vega_datasets import data
state_dat.columns = ['id' if x == 'vega_state_id' else x for x in state_dat.columns]

states = alt.topo_feature(data.us_10m.url, 'states')
variable_list = ['rpt_qnt','inj_qnt','kill_qnt']

alt.Chart(states).mark_geoshape().encode(
    alt.Color(alt.repeat('row'), type='quantitative')
).transform_lookup(
    lookup='id',
    from_=alt.LookupData(state_dat, 'id', variable_list)
).properties(
    width=500,
    height=300
).project(
    type='albersUsa'
).repeat(
    row=variable_list
).resolve_scale(
    color='independent'
)
```

```python
gBy_ltype= law_tbl.groupby(['Type', 'state_abr']).count()
gBy_ltype.reset_index(inplace=True)
gBy_ltype.columns = ['Law_Count' if x=='LawID' else x for x in gBy_ltype.columns]
gBy_ltype = gBy_ltype.drop(['Name', 'Desc'], 1)
gBy_ltype
```

```python
gBy_law = gBy_ltype.groupby(['state_abr'])['Law_Count'].sum()
gBy_law = gBy_law.reset_index()
state_dat = pd.merge(state_dat, gBy_law, how='left', on='state_abr')
state_dat =  state_dat.fillna(0)
state_dat
```

```python
alt.Chart(state_dat).mark_bar().encode(
    alt.X('state_abr:N', sort='-y'),
    y='Law_Count:Q',
    color = 'Type:N'
)
```
